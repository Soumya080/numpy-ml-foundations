MATHS INVOLVED

Input:
X -> shape (100, 1)
y -> shape (100, 1)

Parameters:
w -> shape (1, 1)
b -> shape (1,)

Forward Pass:
ŷ = X @ w + b
ŷ_shape -> (100, 1)

Loss:
MSE = mean((y - ŷ)^2)
Loss_shape -> scalar


#CODE PLAN
1. create a synthetic data(x,y)
2. initialize w and b randomly
3. predict y
4. compute mse loss
5. print loss value


```python
import numpy as np
x= np.random.rand(100,1)
y= 2.5*x +1.0 +0.2*np.random.randn(100,1)
```

```python
w= np.random.randn(1,1)
b= np.random.randn(1)

```

```python
y_pred = x@w + b
```

```python
loss =(( y_pred -y)**2)

print("LOSS")
print(loss.mean())
```
