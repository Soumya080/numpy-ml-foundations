1️⃣ Make predictions  → What do I think the answer is?
2️⃣ Measure loss     → How wrong am I?
3️⃣ Compute gradient  → Which direction reduces the error?
4️⃣ Update params    → Move a little to improve
5️⃣ Repeat           → Learn gradually


```python
import numpy as np
import matplotlib.pyplot as plt

# creating the dataset
x = np.random.randn(100,1)
y = 2.5*x + 0.2*np.random.randn(100,1)

# initializing parameters
w = np.random.randn(1,1)
b = np.random.randn(1,1)

lr = 0.01

loss_history = []

# training loop
for i in range(100):  # 100 iterations
    y_pred = x@w + b
    loss = np.mean((y_pred - y)**2)

    grad_w = np.mean(2*x*(y_pred - y))
    grad_b = np.mean(2*(y_pred - y))

    w -= lr * grad_w
    b -= lr * grad_b
    loss_history.append(loss)

    if i % 10 == 0:
        print(f"Iteration {i+10}: loss = {loss}")

print("Final weights:", w, "Final bias:", b)
print("Final loss:", loss)
# print("final loss history ", loss_history)

x_axis = range(1, len(loss_history) + 1)
plt.plot(x_axis, loss_history)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs Iteration")
plt.show()
```
